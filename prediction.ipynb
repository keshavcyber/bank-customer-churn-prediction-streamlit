```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
import pickle
```

```python
# load pickle file
model=load_model('model.h5')
scaler=pickle.load(open('scaler.pkl','rb')) #either way is fine
'''with open ('scaler.pkl','rb') as f:
    scaler=pickle.load(f)'''
with open('onehotencoder.pkl','rb') as f:
    ohe=pickle.load(f)
with open ('labelencoder.pkl','rb') as f:
    le=pickle.load(f)
```

```python
#example of input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
'''y={
  'Age': [40, 41, 42],
  'Geography': ['France', 'Spain', 'Germany']
}''' #to remember how to create dataframe keys are column in dict that why [] for row

```

```python
#this was my attempt
# preprocess i/p data
x=pd.DataFrame([input_data])
geo=ohe.transform(x[['Geography']]).toarray()
geo_dp=pd.DataFrame(geo,columns=ohe.get_feature_names_out(['Geography']))

geo_dp
```

```python
#lecture way
geo_encoded=ohe.transform([[input_data['Geography']]]).toarray()
df=pd.DataFrame(geo_encoded,columns=ohe.get_feature_names_out(['Geography']))
df
```

```python
input_df=pd.DataFrame([input_data])
input_df['Gender']=le.transform(input_df['Gender'])
```

```python
input_df=pd.concat([input_df.drop(['Geography'],axis=1),geo_dp],axis=1)
input_df
```

```python
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
# model prediction
prediction=model.predict(input_scaled)
prediction
```

```python
prediction_probability=prediction[0][0]
prediction_probability
if prediction_probability>0.5:
    print(f'The customer will leave the bank with a probability of {prediction_probability}')
else:
    print(f'The customer will stay in the bank with a probability of {1-prediction_probability}')
```
